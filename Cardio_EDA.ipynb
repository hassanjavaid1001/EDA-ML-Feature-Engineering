```python
# Commented out IPython magic to ensure Python compatibility.
# 1. Load libraries and data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Display plots inline
# %matplotlib inline

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/Dataset/cardio_train.csv", sep=';')  # often uses semicolon delimiter
df.head()
```

```python
# Check basic information
df.info()
```

```python
# Check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python

# Basic statistics
df.describe()
```

```python
# Countplot for target variable
sns.countplot(x='cardio', data=df)
plt.title("Distribution of Cardiovascular Disease")
plt.xlabel("Cardio (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Boxplot of age vs cardio
sns.boxplot(x='cardio', y='age', data=df)
plt.title("Age vs Cardio")
plt.show()
```

```python
# Pairplot for a few features
sns.pairplot(df[['age', 'weight', 'ap_hi', 'ap_lo', 'cardio']], hue='cardio', diag_kind='kde')
plt.show()
```

```python
print(f"Number of duplicated rows: {df.duplicated().sum()}")

print("Shape:", df.shape)
print("Memory usage (MB):", round(df.memory_usage(deep=True).sum() / 1024**2, 2))
df.info()
```

```python
numeric_cols = ['age', 'height', 'weight', 'ap_hi', 'ap_lo']
df[numeric_cols].hist(figsize=(14, 8), bins=30, edgecolor='black')
plt.tight_layout()
plt.show()
```

```python

correlations = df.corr()['cardio'].drop('cardio').sort_values(ascending=False)
print("Top correlated features:\n", correlations)
```

```python
# Plot
plt.figure(figsize=(8, 4))
sns.barplot(x=correlations.values, y=correlations.index)
plt.title("Feature Correlation with Cardio")
plt.show()
```

```python
grouped = df.groupby('cardio').mean().T
grouped.plot(kind='barh', figsize=(10, 8))
plt.title("Mean Feature Values by Cardio Group")
plt.xlabel("Mean Value")
plt.ylabel("Features")
plt.grid(True)
plt.tight_layout()
plt.show()
```
